```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('tidy_crime_data_sample.csv', header=[0, 1])
df.head()
```

# Categorical variables

```python
for c in [c for c in df.columns if c[0]=='Categorical']:
    sns.countplot(data=df, y=c, order=df[c].value_counts().index)
    plt.figure()
```
