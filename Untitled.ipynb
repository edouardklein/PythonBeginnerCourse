```python
ohe = sklearn.preprocessing.OneHotEncoder(sparse=False, handle_unknown='ignore')
ohe.fit(np.array(sorted(sum([list(df[h].dropna()) for h in ['Infraction 1', 'Infraction 2', 'Infraction 3', 'Infraction 4']], []))).reshape(-1, 1))
```

```python
offense_multiindex=pd.MultiIndex.from_product([['Offense'], ohe.categories_[0]])
offense_tidy_data = sum([
    ohe.transform(np.array(df[colname].apply(lambda x: x if x==x else "N/A")).reshape(-1, 1)) 
        for colname in ['Infraction 1', 'Infraction 2', 'Infraction 3', 'Infraction 4']])
df2 = pd.DataFrame(
    data=offense_tidy_data,
    columns=offense_multiindex, index=df.index)
df2
```
